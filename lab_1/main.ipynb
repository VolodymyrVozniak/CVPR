Call the function **main** (with path for saving images and videos and resolution for video as a parameters) to run the program.

While the program is runnung you may use such buttons:
- **SPACE** - create an image with line and rectangle with random shape and color;
- **R** - start the video with random lines and rectangles;
- **Q** - end the video;
- **ESC** - quit the program **(double press)**.

**Note: you must use only English layout!**

```python
def main(path, resolution):
    
    """
    parameters:
    path - string (path for saving images and videos)
    resolution - tuple (resolution for video)
    
    return value:
    None
    """

    import cv2
    import random as r

    cam = cv2.VideoCapture(0)
    img_counter = 0
    video_counter = 0

    while True:
        ret, frame = cam.read()
        if not ret:
            print("Failed to grab frame")
            break
        cv2.imshow("main_frame", frame)

        k = cv2.waitKey(1)

        if k % 256 == 27:  # if user press ESC
            print("Closing...")
            break

        elif k % 256 == 32:  # if user press SPACE
            # Save img to file
            img_counter += 1
            img_name = path + f"\\image_{img_counter}.png"
            cv2.imwrite(img_name, frame)
            
            # Read the saved img and open in a new window
            img = cv2.imread(img_name, 0)

            # Define random values for colors and size of line and rectangle
            height, width = img.shape
            line_color, rect_color = tuple(r.randint(0, 255) for i in range(3)), tuple(r.randint(0, 255) for i in range(3))
            line_sp, line_ep = (r.randint(0, width-1), r.randint(0, height-1)), (r.randint(0, width-1), r.randint(0, height-1))
            rect_sp, rect_ep = (r.randint(0, width-1), r.randint(0, height-1)), (r.randint(0, width-1), r.randint(0, height-1))

            # Convert image to BGR, so that geometrical figures are coloured
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
            img = cv2.line(img, line_sp, line_ep, line_color, 10)
            img = cv2.rectangle(img, rect_sp, rect_ep, rect_color, 10)
            cv2.namedWindow(img_name, cv2.WINDOW_AUTOSIZE)
            cv2.imshow(img_name, img)

            print("Succssefully created new image!")

        elif k & 0xFF == ord('r'):  # if user press R
            # Define the codec and create VideoWriter object
            fourcc = cv2.VideoWriter_fourcc(*'DIVX')
            video_counter += 1
            out = cv2.VideoWriter(path + f'\\video_{video_counter}.avi', fourcc, 20.0, resolution)

            print("Started to record new video")

            while (cam.isOpened()):
                ret, frame = cam.read()
                if not ret:
                    print("Failed to grab frame")
                    break
                else:
                    # Define random values for colors and size of line and rectangle
                    width, height = resolution
                    line_color, rect_color = tuple(r.randint(0, 255) for i in range(3)), tuple(r.randint(0, 255) for i in range(3))
                    line_sp, line_ep = (r.randint(0, width-1), r.randint(0, height-1)), (r.randint(0, width-1), r.randint(0, height-1))
                    rect_sp, rect_ep = (r.randint(0, width-1), r.randint(0, height-1)), (r.randint(0, width-1), r.randint(0, height-1))

                    # Draw line and rectangle and update the video
                    frame = cv2.line(frame, line_sp, line_ep, line_color, 10)
                    frame = cv2.rectangle(frame, rect_sp, rect_ep, rect_color, 10)
                    out.write(frame)
                    cv2.imshow('main_frame', frame)

                    if cv2.waitKey(1) & 0xFF == ord('q'):  # if user press Q
                        print("Successfully created new video!")
                        break

    cam.release()
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
path = "Storage"
resolution = (640, 480)
main(path, resolution)
```
